### 3D Plot

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
np.random.seed(0)
n_points = 100
x = np.linspace(-5, 5, n_points)
y = np.linspace(-5, 5, n_points)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 - Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of Function')

# Show the 3D plot
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image using OpenCV
image = cv2.imread('AIR Logo.png')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to create a blurred version of the image
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Calculate the unsharp mask by subtracting the blurred image from the original grayscale image
unsharp_mask = cv2.subtract(gray_image, blurred_image) * 2

# Adjust the strength of the sharpening effect by scaling the unsharp mask
sharpened_image = cv2.addWeighted(gray_image, 1.5, unsharp_mask, -0.5, 0)

# Display the original and sharpened images using Matplotlib
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title('Original Image')
axes[1].imshow(sharpened_image, cmap='gray')
axes[1].set_title('Sharpened Image')
plt.show()
````

````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generate a time array
t = np.linspace(0, 1, 1000)

# Generate a signal with two frequencies
frequency1 = 100
frequency2 = 250
signal_wave = np.sin(2 * np.pi * frequency1 * t) + np.sin(2 * np.pi * frequency2 * t)
# Plot the signal
plt.plot(t, signal_wave)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()

# Compute the FFT
fft = np.fft.fft(signal_wave)

# Compute the frequencies corresponding to the FFT coefficients
frequencies = np.fft.fftfreq(len(signal_wave), t[1] - t[0])

# Plot the FFT spectrum
plt.plot(frequencies, np.abs(fft))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT Spectrum')
plt.show()
````

````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generate a signal with noise
t = np.linspace(0, 1, 1000)
signal_wave = np.sin(2 * np.pi * 10 * t) + np.random.normal(0, 0.5, len(t))

# Define the window for the moving average filter
window = signal.windows.hann(50)

# Apply the moving average filter
filtered_signal = signal.convolve(signal_wave, window, mode='same') / sum(window)

# Plot the original and filtered signals
plt.plot(t, signal_wave, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Moving Average Filter')
plt.legend()
plt.show()
````

````python
import numpy as np

def calculate_correlation(vector1, vector2):
    if len(vector1) != len(vector2):
        raise ValueError("Vectors must have the same length.")

    n = len(vector1)

    # Calculate the means
    mean1 = np.mean(vector1)
    mean2 = np.mean(vector2)

    # Calculate the standard deviations
    std1 = np.std(vector1)
    std2 = np.std(vector2)

    # Calculate the covariance
    covariance = np.sum((vector1 - mean1) * (vector2 - mean2)) / n

    # Calculate the correlation coefficient
    correlation = covariance / (std1 * std2)

    return correlation

# Example usage
vector1 = [1,2,3,4,5]
vector2 = [8,500,7,95,-98]

correlation = calculate_correlation(vector1, vector2)
print("Correlation coefficient:", correlation)
````

# Pandas Data Series
* A Pandas Data Series is a one-dimensional labeled array capable of holding any data type (integers, floats, strings, Python objects, etc.). It is similar to a column in a spreadsheet or a list in Python but with additional functionality provided by Pandas.

Key Features of Pandas Series:
* **_Labeled Index_**  Each element in the series has an associated index (by default, it starts from 0).
* **_Homogeneous Data_**  It stores data of one type (e.g., all integers, all floats, etc.).
* **_Supports Various Data Sources_**  A Pandas Series can be created from lists, dictionaries, NumPy arrays, etc.
* **_Vectorized Operations_**  Supports fast mathematical operations like addition, subtraction, filtering, etc.

## Default Numeric Type in Pandas:

* If all values are integers, Pandas assigns int64 as the default.
* If all values are floating-point numbers, Pandas assigns float64.
* If mixed types are present (e.g., string and number), Pandas assigns object.

````python
import pandas as pd

# Create a DataFrame
data = {'Name': ['John', 'Jane', 'Mike', 'Emily'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'London', 'Paris', 'Sydney']}
````

````python
df = pd.DataFrame(data)
````

````python
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])
````

````python
df
````

````python
# Accessing rows using loc
row_loc = df.loc['A']
print("Row using loc:\n", row_loc)
````

````python
# Accessing rows using iloc
row_iloc = df.iloc[0] #Uses integer positions (0-based index) to select rows/columns.
print("Row using iloc:\n", row_iloc)
````

````python
# Accessing specific value using loc
value_loc = df.loc['A', 'Age']
print("Value using loc:", value_loc)
````

````python
# Accessing specific value using iloc
value_iloc = df.iloc[1, 1]
print("Value using iloc:", value_iloc)
````

# Scipy


````python

from scipy import linalg

import numpy as np
numpy_mat = np.array([[3,2,2],[5,5,2],[7,8,69]])
linalg.det(numpy_mat)
````

````python
EW, EV = linalg.eig(numpy_mat)
print(EW)
print(EV)
````

````python
#Computation of FFT
from scipy.fftpack import fft,ifft
import numpy as np
#Numpy array generation
x = np.array([1.0, 1.0, 1.0, 1.0])

#Applying the fft function
y = fft(x)
print(y)
````

````python
yinv = ifft(y)

print (yinv)
````
